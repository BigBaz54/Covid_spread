# Covid spread relation with position

In this first notebook, we'll compare the spread of Covid-19 with the position of the countries. We'll use the latitude and longitude of the countries to see if there is any relation between the position of the countries and the spread of the virus.

The goal is to reproduce the results of the Kaggle article [Does latitude impact the spread of COVID-19?](https://www.kaggle.com/code/paultimothymooney/does-latitude-impact-the-spread-of-covid-19/notebook) that inspired this project.

```python
import pandas as pd
import matplotlib.pyplot as plt


data = pd.read_csv('data/data_spread_rate.csv')
```

## Mortality rate vs. absolute latitude
The original article actually only use the mortality rate to compare the spread of the virus with the position of the countries. It claims that the mortality rate is higher in countries (or US states) with higher absolute latitude.

```python
plt.figure(figsize=(10, 10))
plt.rcParams.update({'font.size': 15})
latitude = abs(data['Latitude'])
plt.scatter(latitude, data['Mortality Rate'], c='blue', alpha=0.5)
plt.xlabel('Absolute Latitude')
plt.ylabel('Mortality Rate')
plt.title('Mortality Rate vs Absolute Latitude')
plt.show()
```

```python
correlation = data['Mortality Rate'].corr(latitude)
print('Correlation between Mortality Rate and Absolute Latitude:', correlation)
```

We used the mortality rate from the original article to plot the same graph. (The only difference is that few countries have been removed from the original dataset because of missing data.)

The graph and the correlation shows a weak correlation between the mortality rate and the absolute latitude.

## Spread rate vs. absolute latitude
We computed our own spread rate for each country as explained in the corresponding notebook. We can then plot the spread rate against the absolute latitude of the countries to see if there is any correlation.

```python
plt.figure(figsize=(10, 10))
plt.rcParams.update({'font.size': 15})
latitude = abs(data['Latitude'])
plt.scatter(latitude, data['Spread rate'], c='blue', alpha=0.5)
plt.xlabel('Absolute Latitude')
plt.ylabel('Spread rate')
plt.title('Spread rate vs Absolute Latitude')
plt.show()
```

```python
correlation = data['Spread rate'].corr(latitude)
print('Correlation between Spread rate and Absolute Latitude:', correlation)
```

Using our spread rate, the correlation with the absolute latitude is even weaker than with the mortality rate.

## Spread rate vs. longitude
We can also plot the spread rate against the longitude of the countries to see if there is any correlation.

```python
plt.figure(figsize=(10, 10))
plt.rcParams.update({'font.size': 15})
longitude = data['Longitude']
plt.scatter(latitude, data['Spread rate'], c='blue', alpha=0.5)
plt.xlabel('Longitude')
plt.ylabel('Spread rate')
plt.title('Spread rate vs Longitude')
plt.show()
```

```python
correlation = data['Spread rate'].corr(longitude)
print('Correlation between Spread rate and Longitude:', correlation)
```

The correlation with the longitude is as weak as with the latitude.

## Conclusion
The original article claims that the mortality rate is higher in countries with higher absolute latitude. We reproduced the results and found a weak correlation between the mortality rate and the absolute latitude.

We also computed our own spread rate and found a weak correlation with the absolute latitude and the longitude.

The results confirm the intuition that the spread of the virus is not related to the position of the countries.
